```python
with open("labeled_reviews.txt", 'r') as f:
    n_tag_classes = DEFAULT_N_CLASSES
    tag_vector_map = {}
    tag_class_id = 0
    raw_news_data = []
    raw_news_words = []
    raw_news_tags = []  
    for line in f:
        line = line.strip()
        if not line:
            raw_news_data.append( (tuple(raw_news_words), tuple(raw_news_tags)) )
            raw_news_words = []
            raw_news_tags = []
            continue
        word, tag = line.split('\t')
        raw_news_words.append(word)
        raw_news_tags.append(tag)
        if tag not in tag_vector_map:
            print(tag)
            one_hot_vec = np.zeros(DEFAULT_N_CLASSES, dtype=np.int32)
            one_hot_vec[tag_class_id] = 1
            tag_vector_map[tag] = tuple(one_hot_vec)
            tag_vector_map[tuple(one_hot_vec)] = tag
            tag_class_id += 1
```

```python
all_X,all_Y = [], []
unk_words = []
for words, tags in raw_news_data:
    elem_wordvecs, elem_tags = [], []
    for ix in range(len(words)):
        w = words[ix]
        t = tags[ix]
        if w in word_to_ix_map:
            elem_wordvecs.append(wordvecs[word_to_ix_map[w]])
            elem_tags.append(list(tag_vector_map[t]))
        elif False:
            unk_words.append(w)
            continue
        else:
            unk_words.append(w)
            new_wv = 2*np.random.randn(300)-1 # sample from normal distn
            norm_const = np.linalg.norm(new_wv)
            new_wv /= norm_const
            word_to_ix_map[w] = wordvecs.shape[0]
            print(wordvecs.shape[0])
            wordvecs = np.vstack((wordvecs, new_wv))
            elem_wordvecs.append(new_wv)
            elem_tags.append(list(tag_vector_map[t]))
    nil_X = np.zeros(300)
    nil_Y = np.array(tag_vector_map['NIL'])
    pad_length = max_sentence_len - len(elem_wordvecs)
    all_X.append( ((pad_length)*[nil_X]) + elem_wordvecs)
    all_Y.append( ((pad_length)*[nil_Y]) + elem_tags)
```

```python
test_split_mask = np.random.rand(len(all_X)) < (1-0.2)
```

```python
all_X=np.array(all_X)
all_Y=np.array(all_Y)
```

```python
train_X = all_X[test_split_mask]
train_Y= all_Y[test_split_mask]
test_X = all_X[~test_split_mask]
test_Y= all_Y[~test_split_mask]
```

```python
model = Sequential()
reg_alpha = 0.000
dropout = 0.5

model.add(Bidirectional(LSTM(150, return_sequences=True, \
                                        W_regularizer=l2(reg_alpha), \
                                        U_regularizer=l2(reg_alpha), \
                                        b_regularizer=l2(reg_alpha)), \
                                input_shape=(50,300)))
model.add(Dropout(dropout))
if 1 > 1:
    model.add(Bidirectional(LSTM(150, return_sequences=True, \
                                            W_regularizer=l2(reg_alpha), \
                                            U_regularizer=l2(reg_alpha), \
                                            b_regularizer=l2(reg_alpha))))
    
    model.add(Dropout(dropout))
model.add(TimeDistributed(Dense(6, activation='softmax', \
                                        W_regularizer=l2(reg_alpha), \
                                        b_regularizer=l2(reg_alpha))))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
model.fit(train_X, train_Y, nb_epoch=10, batch_size=50)
```

```python
scores = model.evaluate(test_X, test_Y, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```

```python
def encode_sentence(sentence):
    X = []
    outer_X = []
    for word in sentence:
        if word in word_to_ix_map:
            X.append(wordvecs[ word_to_ix_map[word] ])
        else:
            new_wv = 2*np.random.randn(300)-1 # sample from normal distn
            norm_const = np.linalg.norm(new_wv)
            new_wv /= norm_const
            X.append(new_wv)        
    outer_X.append(X)
    return sequence.pad_sequences(outer_X, maxlen=max_sentence_len, dtype=np.float64)
```

```python
def decode_prediction_sequence (pred_seq):
    pred_tags = []
    for class_prs in pred_seq:
        class_vec = np.zeros(6, dtype=np.int32)
        class_vec[ np.argmax(class_prs) ] = 1
        if tuple(class_vec.tolist()) in tag_vector_map:
            pred_tags.append(tag_vector_map[tuple(class_vec.tolist())])
    return pred_tags

```

```python
def predict_sentence(sentence):
    sen=sentence[:51]
    x=encode_sentence(sen)
    pred=model.predict(x,batch_size=1)
    tags=decode_prediction_sequence(pred[0])
    tag_str = ""
    if not False:
        tags = tags[-len(sen):]
    for t in tags:
        tag_str += t + " "
    return tags
```

```python
def judge(labe):
    post=[]
    i=0
    while i <len(labe):
        if labe[i]=='U':
            post.append([i])
            i=i+1
        elif i<(len(labe)-1) and labe[i]=='B' and labe[i+1]=='L':
            post.append([i,i+1])
            i=i+2
        elif i<(len(labe)-1) and labe[i]=='B' and labe[i+1]!='L':
            temp=[i]
            i=i+1
            while i<(len(labe)) and labe[i]=='I':
                temp.append(i)
                i=i+1
            if i<(len(labe)) and labe[i]=='L':
                temp.append(i)
                post.append(temp)
        else:
            i=i+1
    return post
        
```
