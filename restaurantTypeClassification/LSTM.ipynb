```python
import pandas as pd
```

```python
businessFrame = pd.read_csv("business.csv", index_col=False)
reviewFrame = pd.read_csv("review.csv", index_col=False)
```

```python
businessFrame = businessFrame[['categories', 'business_id']]
reviewFrame = reviewFrame[['text', 'business_id', 'useful']]
```

```python
tag=["Food","Restaurants","Fast Food","Bars","Coffee & Tea"]
```

```python
b_row = businessFrame.shape[0]
```

```python
isFood = []
for i in range(0,b_row):
    isFood.append(any(t in tag for t in eval(businessFrame.iloc[:,0][i])))
```

```python
df_food = businessFrame[isFood].copy()
```

```python
b_row = df_food.shape[0]
print(b_row)
```

```python
df_food.index = range(b_row)
```

```python
label_set = []
for i in range(0,b_row):
    for t in eval(df_food.iloc[:,0][i]):
        label_set.append(t)
label_set = label_set
```

```python
import collections
freqdict=collections.Counter(label_set)
```

```python
import numpy as np
```

```python
def find_the_max(list1,diction):
    count=0
    item="CANCEL"
    for j in list1:
        if j!="Food" and j!="Restaurants" and j!="Nightlife":
            if diction[j]>count:
                count=diction[j]
                if j=="American (Traditional)" or j=="American (New)":
                    j="American"
                item=j
    return item
```

```python
truelabel=[]
for i in range(b_row):
    truelabel.append(find_the_max(eval(df_food.iloc[:,0][i]),freqdict))
```

```python
df_food["Truelabel"]=truelabel
```

```python
df_food
```

```python
def find_country(dataf,name):
    count=[]
    for k in range(dataf.shape[0]):
        count.append(dataf.iloc[k,2]==name)
    return count
```

```python
aa= np.logical_or(find_country(df_food,"American"),find_country(df_food,"Chinese"))
bb=np.logical_or(find_country(df_food,"Italian"),find_country(df_food,"Japanese"))
cc=np.logical_or(find_country(df_food,"Mexican"),find_country(df_food,"Indian"))
dd=np.logical_or(find_country(df_food,"Thai"),find_country(df_food,"French"))
final1=np.logical_or(aa,bb)
final2=np.logical_or(cc,dd)
final=np.logical_or(final1,final2)
```

```python
dffinal_food=df_food[final].copy()
```

```python
dffinal_food.index = range(dffinal_food.shape[0])
```

```python
finalnewdf=pd.merge(dffinal_food, reviewFrame, on='business_id').iloc[:,2:5].copy()
```

```python
dfff=finalnewdf[finalnewdf.iloc[:,2]>1]
```

```python
dfff.index = range(dfff.shape[0])
```

```python
dfff
```

```python
labels = ["American", "Mexican", "Chinese", "Japanese", "Thai", "Italian", "Indian", "French"]
dfs = []
dft = []
for label in labels:
    kk=int(np.sum(dfff.iloc[:,0]==label)*0.8)
    dfs.append(dfff[dfff.iloc[:,0]==label][0:kk])
    dft.append(dfff[dfff.iloc[:,0]==label][kk:])
pdd=pd.concat(dfs)
pdt=pd.concat(dft)
```

```python
pdd=pdd.sort_index().copy()
pdd.index = range(pdd.shape[0])
```

```python
pdt=pdt.sort_index().copy()
pdt.index = range(pdt.shape[0])
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
MAX_NB_WORDS = 4000
tokenizer = Tokenizer(num_words=MAX_NB_WORDS)
```

```python
review_train = [string[1:].replace("\\n", "") for string in pdd["text"]]
```

```python
review_test = [string[1:].replace("\\n", "") for string in pdt["text"]]
```

```python
MAX_SEQUENCE_LENGTH = 200
fit_texts = review_train + review_test
tokenizer.fit_on_texts(fit_texts)
sequences = tokenizer.texts_to_sequences(review_train)
data_train = pad_sequences(sequences, maxlen=MAX_SEQUENCE_LENGTH)
sequences = tokenizer.texts_to_sequences(review_test)
data_test = pad_sequences(sequences, maxlen=MAX_SEQUENCE_LENGTH)
```

```python
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
# encode class values as integers
Y = pdd["Truelabel"]
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
# convert integers to dummy variables (i.e. one hot encoded)
y_train = np_utils.to_categorical(encoded_Y)
```

```python
# encode class values as integers
Y = pdt["Truelabel"]
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
y_test = np_utils.to_categorical(encoded_Y)
```

```python
x_train=data_train
x_test=data_test
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence
```

```python
embedding_vecor_length = 32
model = Sequential()
model.add(Embedding(MAX_NB_WORDS, embedding_vecor_length, input_length=MAX_SEQUENCE_LENGTH))
model.add(LSTM(100))
model.add(Dense(8, input_dim=4, activation='relu'))
model.add(Dense(8, activation='softmax'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['top_k_categorical_accuracy'])
print(model.summary())
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=3, batch_size=64)
```

```python
scores = model.evaluate(x_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```

```python
labels = list(encoder.classes_)
```

```python
sequences = tokenizer.texts_to_sequences([review_test[2000]])
test_x = pad_sequences(sequences, maxlen=MAX_SEQUENCE_LENGTH)
y_predicted = model.predict(test_x)
zipped = zip(labels, y_predicted[0])
print(sorted(zipped, key=lambda elem: elem[1], reverse=True))
```
