```python
tag=["Food","Restaurants","Fast Food","Bars","Coffee & Tea"]
```

```python
#Find out the categories that are relevant to food according to the tag
def find_food(business_Frame,Tag):
    isFood = []
    b_row = businessFrame.shape[0]
    for i in range(0,b_row):
        isFood.append(any(t in Tag for t in eval(business_Frame.iloc[:,0][i])))
    df_food = business_Frame[isFood].copy()
    return df_food
```

```python
food_all = find_food(businessFrame,tag)
```

```python
#Catch all food labels
def catch_label(foodAll):
    b_row = foodAll.shape[0]
    foodAll.index = range(b_row)
    label_set = []
    for i in range(0,b_row):
        for t in eval(foodAll.iloc[:,0][i]):
            label_set.append(t)
    return label_set
```

```python
label_all=catch_label(food_all)
```

```python
import numpy as np
```

```python
#Find the maximum probability label to be the true label
def find_the_max(list1,diction):
    count=0
    item="CANCEL"
    for j in list1:
        if j!="Food" and j!="Restaurants" and j!="Nightlife":
            if diction[j]>count:
                count=diction[j]
                if j=="American (Traditional)" or j=="American (New)":
                    j="American"
                item=j
    return item
```

```python
#find the true label
def find_truelabel(foodAll,freq_Dict):
    true_label=[]
    b_row = foodAll.shape[0]
    for i in range(b_row):
        true_label.append(find_the_max(eval(foodAll.iloc[:,0][i]),freq_Dict))
    return true_label
```

```python
truelabel=find_truelabel(food_all,freqdict)
```

```python
def find_country(dataf,name):
    count=[]
    for k in range(dataf.shape[0]):
        count.append(dataf.iloc[k,2]==name)
    return count
```

```python
#Label each review to a country
def label_country(foodAll,pD):
    aa=np.logical_or(find_country(food_all,"American"),find_country(food_all,"Chinese"))
    bb=np.logical_or(find_country(food_all,"Italian"),find_country(food_all,"Japanese"))
    cc=np.logical_or(find_country(food_all,"Mexican"),find_country(food_all,"Indian"))
    dd=np.logical_or(find_country(food_all,"Thai"),find_country(food_all,"French"))
    final1=np.logical_or(aa,bb)
    final2=np.logical_or(cc,dd)
    final=np.logical_or(final1,final2)
    
    dffinal_food=food_all[final].copy()
    dffinal_food.index = range(dffinal_food.shape[0])
    finalnewdf=pD.merge(dffinal_food, reviewFrame, on='business_id').iloc[:,2:5].copy()
    dfff=finalnewdf[finalnewdf.iloc[:,2]>1]
    dfff.index = range(dfff.shape[0])
    return dfff
```

```python
labelCountry=label_country(food_all,pd)
labels = ["American", "Mexican", "Chinese", "Japanese", "Thai", "Italian", "Indian", "French"]
```

```python
#Divide the total set into training and testing set
def divide_set(labelcountry,pD):
    dfs = []
    dft = []
    for label in labels:
        kk=int(np.sum(labelcountry.iloc[:,0]==label)*0.8)
        dfs.append(labelcountry[labelcountry.iloc[:,0]==label][0:kk])
        dft.append(labelcountry[labelcountry.iloc[:,0]==label][kk:])
    pdd=pD.concat(dfs)
    pdt=pD.concat(dft)
    pdd=pdd.sort_index().copy()
    pdd.index = range(pdd.shape[0])
    pdt=pdt.sort_index().copy()
    pdt.index = range(pdt.shape[0])
    return pdd,pdt
```

```python
#Transfer counts into log results
def normalize_counts(counts):
    total = sum(counts.values())
    lg_total = math.log(total/10000)
    return dict((key, math.log(cnt)-lg_total) for key, cnt in counts.items())   
```

```python
#Get words that appear more than 100 times in all reviews and remove others
def to_dictionary(indexs,diction):
    dic={}
    for i in labels:
        dic[i]={}
        for j in list(indexs[i].keys()):
            if diction[j]>100:
                dic[i][j]=indexs[i][j]
        dic[i]['UNK']=0.00001
    return dic
```

```python
#count the number of each word in all reviews and put in a dictionary
def to_dic(indexs):
    dic={}
    for i in labels:
        for j in list(indexs[i].keys()):
            if j in dic.keys():
                dic[j]=dic[j]+indexs[i][j]
            else:
                dic[j]=indexs[i][j]
    return dic
```

```python
#Tokenize a review
def words(text):
    temp=word_tokenize(text)
    word_num=[w.lower() for w in temp if w.isalpha()]
    word_num=word_num[1:]
    return word_num
```

```python
#Use tokenizor to get NN words and put the results in "speech_counts"
speech_counts=[]
for j in total_counts:
    tokenized = nltk.word_tokenize(j)
    tagged=nltk.pos_tag(tokenized)
    if tagged[0][1]=='NN':
        if tagged[0][0]!='i':
            speech_counts.append(tagged[0][0])
```

```python
#Calculate the word counts to log, the results are p(word|category) 
word_given_cat_prob = {}
for cat, counts in word_counts.items():
    word_given_cat_prob[cat] = normalize_counts(counts)

# Calculate probability of each category based on review number. The result if p(category)
seen_categories = set(word_counts)
seen_category_counts = dict((cat, count) for cat, count in category_counts.items())
category_prob = normalize_counts(seen_category_counts)
```

```python
#Divide each review into words and get NN words from them
def noun(review):
    part_speech=[]
    a=words(review)
    for j in a:
        if j in speech_counts:
            part_speech.append(j)
    return part_speech
```

```python
#The classifier for Naive Bayes to classify the review
def classifier(text):
    
#load the word count dictionary
    lg_scores=category_prob.copy()

#Divide each review into words and get NN words from them
    a=noun(text)

#Calculte p(word|category)*p(category). Because each count is transferred to log, all the results are added together
    for word in a:
        for cat in lg_scores:
            cat_probs=word_given_cat_prob[cat]
            if word in cat_probs:
                #find the word
                lg_scores[cat]+=cat_probs[word]
            else:
                #Word not found, the word is recognized as 'UNK'
                lg_scores[cat] += cat_probs['UNK']
    
    #Transfer the result from log to original value
    scores=dict((cat, math.exp(score)) for cat, score in lg_scores.items())
    #Calculate the probability of each category
    total=sum(scores.values())
    return dict((cat, prob/total) for cat, prob in scores.items()) 
```

```python
#Check if the label provided by the classifier is the same as true label. Return 1: same. Result 0: Not
def maximum(text,country):
    guesses=classifier(text)
    max=-1000
    gue=''
    for guess, prob in guesses.items():
        if prob>max:
            max=prob
            gue=guess
    if gue==country:
        return [1,guesses['American'],guesses['Chinese'],guesses['French'],guesses['Indian'],guesses['Italian'],guesses['Japanese'],guesses['Mexican'],guesses['Thai']]
    else:
        return [0,guesses['American'],guesses['Chinese'],guesses['French'],guesses['Indian'],guesses['Italian'],guesses['Japanese'],guesses['Mexican'],guesses['Thai']]     
```
